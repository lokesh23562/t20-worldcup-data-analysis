```python
import pandas as pd

# ---------- Clean Batting Data ----------
# Extract only numeric part of Highest Score
df["Highest Score"] = df["Highest Score"].astype(str).str.extract(r"(\d+)")
df["Highest Score"] = pd.to_numeric(df["Highest Score"], errors="coerce")

# Top Run Scorers
top_runs = (
    df.groupby("Top Scorer")["Highest Score"]
    .agg(["sum", "count"])
    .reset_index()
    .rename(columns={"sum": "Total Runs", "count": "Inns"})
)
top_runs["Average"] = (top_runs["Total Runs"] / top_runs["Inns"]).round(2)
top_runs = top_runs.sort_values(by="Total Runs", ascending=False).head(10)

print("üèè Top 10 Run Scorers:")
display(top_runs)


# ---------- Clean Bowling Data ----------
# Extract only wickets from figures like "4/32"
df["Wickets"] = df["Best Bowler Figure(Wickets Taken)"].astype(str).str.extract(r"(\d+)")
df["Wickets"] = pd.to_numeric(df["Wickets"], errors="coerce")

# Top Wicket Takers
top_bowlers = (
    df.groupby("Best Bowler")["Wickets"]
    .agg(["sum", "count"])
    .reset_index()
    .rename(columns={"sum": "Total Wickets", "count": "Matches"})
)
top_bowlers["Average Wkts/Match"] = (top_bowlers["Total Wickets"] / top_bowlers["Matches"]).round(2)
top_bowlers = top_bowlers.sort_values(by="Total Wickets", ascending=False).head(10)

print("üéØ Top 10 Wicket Takers:")
display(top_bowlers)
```

```python
import pandas as pd

# ---------- Clean Batting Data ----------
# Extract only numeric part of Highest Score
df["Highest Score"] = df["Highest Score"].astype(str).str.extract(r"(\d+)")
df["Highest Score"] = pd.to_numeric(df["Highest Score"], errors="coerce")

# Top Run Scorers
top_runs = (
    df.groupby("Top Scorer")
      .agg(Total_Runs=("Highest Score", "sum"), Inns=("Highest Score", "count"))
      .reset_index()
)
top_runs["Average"] = (top_runs["Total_Runs"] / top_runs["Inns"]).round(2)
top_runs = top_runs.sort_values(by="Total_Runs", ascending=False).head(10)

print("üèè Top 10 Run Scorers:")
display(top_runs)


# ---------- Clean Bowling Data ----------
# Extract only wickets from figures like "4/32"
df["Wickets"] = df["Best Bowler Figure(Wickets Taken)"].astype(str).str.extract(r"(\d+)")
df["Wickets"] = pd.to_numeric(df["Wickets"], errors="coerce")

# Top Wicket Takers
top_bowlers = (
    df.groupby("Best Bowler")
      .agg(Total_Wickets=("Wickets", "sum"), Matches=("Wickets", "count"))
      .reset_index()
)
top_bowlers["Average Wkts/Match"] = (top_bowlers["Total_Wickets"] / top_bowlers["Matches"]).round(2)
top_bowlers = top_bowlers.sort_values(by="Total_Wickets", ascending=False).head(10)

print("üéØ Top 10 Wicket Takers:")
display(top_bowlers)
```

```python
# ---------- Venue Analysis ----------

# Most Matches Played at Venues
venue_matches = (
    df["Venue"].value_counts()
      .reset_index()
      .rename(columns={"index": "Venue", "Venue": "Matches"})
      .head(10)
)

print("üèüÔ∏è Top 10 Venues by Matches Played:")
display(venue_matches)


# Average Runs at Venues (using Highest Score as proxy)
venue_runs = (
    df.groupby("Venue")["Highest Score"]
      .mean()
      .reset_index()
      .rename(columns={"Highest Score": "Avg Highest Score"})
      .sort_values(by="Avg Highest Score", ascending=False)
      .head(10)
)

print("üìä Venues with Highest Average Scores:")
display(venue_runs)


# Average Wickets at Venues
venue_wkts = (
    df.groupby("Venue")["Wickets"]
      .mean()
      .reset_index()
      .rename(columns={"Wickets": "Avg Wickets Taken"})
      .sort_values(by="Avg Wickets Taken", ascending=False)
      .head(10)
)

print("üéØ Venues Supporting Bowlers (Highest Avg Wickets):")
display(venue_wkts)
```

```python
# Convert Wickets column to numeric (invalid values become NaN)
df["Best Bowler Figure(Wickets Taken)"] = pd.to_numeric(
    df["Best Bowler Figure(Wickets Taken)"], errors="coerce"
)

# Drop rows where Wickets is NaN (invalid data)
df_clean = df.dropna(subset=["Best Bowler Figure(Wickets Taken)"])

# Now calculate top wicket takers
top_wickets = (
    df_clean.groupby("Best Bowler")["Best Bowler Figure(Wickets Taken)"]
      .sum()
      .reset_index()
      .sort_values(by="Best Bowler Figure(Wickets Taken)", ascending=False)
      .head(10)
)

print("Top 10 Wicket Takers (Cleaned):")
print(top_wickets)

```
